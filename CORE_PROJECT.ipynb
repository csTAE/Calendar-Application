```python
import calendar

def display_month(year, month):
    cal = calendar.TextCalendar(calendar.SUNDAY)
    month_str = cal.formatmonth(year, month)
    print(month_str)
```

```python
from datetime import datetime

events = []

def add_event(date_str, event):
    date_obj = datetime.strptime(date_str, '%Y-%m-%d')
    events.append({'date': date_obj, 'event': event})
    print(f"Event '{event}' added on {date_str}.")

def view_events():
    if not events:
        print("No events found.")
    else:
        for event in events:
            print(event['date'].strftime('%Y-%m-%d'), event['event'])

def delete_event(date_str, event):
    date_obj = datetime.strptime(date_str, '%Y-%m-%d')
    global events
    events = [e for e in events if not (e['date'] == date_obj and e['event'] == event)]
    print(f"Event '{event}' on {date_str} deleted.")
```

```python
def main():
    while True:
        print("\nCalendar Application")
        print("1. Display Monthly Calendar")
        print("2. Add Event")
        print("3. View Events")
        print("4. Delete Event")
        print("5. Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == '1':
            year = int(input("Enter year: "))
            month = int(input("Enter month: "))
            display_month(year, month)
        elif choice == '2':
            date_str = input("Enter date (YYYY-MM-DD): ")
            event = input("Enter event: ")
            add_event(date_str, event)
        elif choice == '3':
            view_events()
        elif choice == '4':
            date_str = input("Enter date (YYYY-MM-DD): ")
            event = input("Enter event: ")
            delete_event(date_str, event)
        elif choice == '5':
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
